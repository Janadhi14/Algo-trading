```python
#setting a variable called aapl and then creating a datafram that contains the close values 
aapl= close.AAPL.copy().to_frame() # the to_frame method will convert the data to a dataframe 
aapl
```

```python
aapl.shift(periods = 1), aapl #this will off set the data by n(periods) which can be usefull for manipulation of offset data 
```

```python
aapl["lag1"] = aapl.shift(periods= 1)
aapl
```

```python
aapl["Diff"]= aapl.AAPL.sub(aapl.lag1)
```

```python
aapl
```

```python
aapl["% Change"]=aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl
# now we are able to see the percentage change 
```

```python
aapl["Diff2"] = aapl.AAPL.diff(periods= 1)
aapl
# another way of getting the difference 
```

```python
#we are also able to use the percentage change function to this function
aapl["% Change2"] = aapl.AAPL.pct_change(periods= 1).mul(100)
aapl
```

```python
del aapl["Diff2"]
aapl
```

```python
aapl.rename(columns = {'% Change2': 'Change'}, inplace = True)
aapl
```

```python
aapl.AAPL.resample("M").last()
```

```python
# we can do last business day of the month using BM as the parameter in resample

aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

```python
aapl.pct_change()
#here we are getting a percentage change 
```

```python
# creating a new variable ret and setting that is going to calculate the daily returns (% change during the day )
ret = aapl.pct_change().dropna() # dropna will remove the na from the data frame 
```

```python
ret.info()
```

```python
# here we are creating a plot of the 
ret.plot(kind= "hist", figsize = (12,8), bins= 100) # plotting a histagram,bins is the bins that your data will be divided into
plt.show
```

- shows that majority of the trading range(percentage of change tends to be between 0.05 and -0.05 percentage change)

```python
# lets calcuate the mean return 
daily_mean_return= ret.mean()
daily_mean_return
```

```python
# lets calculate the varience
var_daily = ret.var() 
var_daily 
```

```python
# lets look at the standard deviation 
st_daily =np.sqrt(var_daily)
st_daily
```

```python
# lets calculate the annual mean return 
annual_mean_return = daily_mean_return*252
annual_mean_return
```

```python
annual_var_return = var_daily*252
annual_var_return
```

```python
# now lets calculate the annual standard return  
annual_standard_return = np.sqrt(annual_var_return)
annual_standard_return 
```

```python
#we can also get the same answer thorugh multplying by np.sqrt(255)
ret.std()*np.sqrt(252)
#same answer as annual_standard_return
```

#### **Cummalive Returns and Drawdowns**

```python
aaple = yf.download("AAPL")
```

```python
aaple= aaple.Close.to_frame()
```

```python
aaple["d_returns"]=np.log(aaple.div(aaple.shift(1)))
```

```python
aaple
```

```python
aaple.dropna(inplace=True)
aaple
```

```python
aaple.d_returns.sum()
```

```python
np.exp(aaple.d_returns.sum())
```

```python
#so for every one dollar that has been invested in aaple in 1980 it would be worth 1170
```

```python
#cummulative returns 
aaple["cummreturns"]=aaple.d_returns.cumsum().apply(np.exp)
```
